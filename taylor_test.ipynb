```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def taylor_trig(x) -> list:
    # taylor series expansion for sin(x) and cos(x)

    sin_x = x - (x**3)/6 + (x**5)/120 - (x**7)/5040 + \
        (x**9)/362880 + (x**11)/39916800 + (x**13)/6227020800 + \
        (x**15)/1307674368000 + (x**17)/355687428096000

    cos_x = 1 - (x**2)/2 + (x**4)/24 - (x**6)/720 + (x**8)/40320 + \
        (x**10)/3628800 + (x**12)/479001600 + (x**14)/87178291200 + \
        (x**16)/20922789888000 + (x**18)/6402373705728000

    return [cos_x, sin_x]

def np_trig(x) -> list:
    # return cos(x) and sin(x)
    return(np.cos(x), np.sin(x))

def osc_plot(x, y, x2, y2):
    plt.style.use('dark_background')
    plt.rcParams['lines.linewidth'] = 0.5
    plt.figure(figsize=(16, 16))
    plt.plot(x, y, 'r', label='Taylor Series Expansion')
    plt.plot(x2, y2, 'b', label='numpy trig functions')
    plt.plot(-x, y, 'r')
    plt.plot(-x2, y2, 'b')
    plt.plot(x, -y, 'r')
    plt.plot(x2, -y2, 'b')
    plt.plot(-x, -y, 'r')
    plt.plot(-x2, -y2, 'b')
    plt.legend(loc='upper left',
        prop={'size': 20})
    plt.title('Unit Circle Approximations', fontsize=30)
    plt.show()
```

```python
# generate 1000 points between -pi and pi
rad_osc = np.linspace(-np.pi, np.pi, 1000)

# generate coordinates for the oscillation
x_t, y_t = taylor_trig(rad_osc)
x_np, y_np = np_trig(rad_osc)

# plot the oscillations
osc_plot(x_t, y_t, x_np, y_np)
```
